## Importing Libraries

```python
import pandas
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
```

```python
trends = TrendReq()
```

Analyzing the Google search trends on the queries based on "Machine Learning". Let's create a DataFrame of top 10 countries which search for "Machine Learning"

```python
trends.build_payload(kw_list=["Machine Learning"])
data = trends.interest_by_region()
data = data.sort_values(by="Machine Learning", ascending=False)
data = data.head(10)
print(data)
```

## Visualizing the Results

```python
data.reset_index().plot(x="geoName", 
                        y="Machine Learning",
                        figsize=(15, 12),
                        kind="bar")
plt.style.use('fivethirtyeight')
plt.show()
```

## Trend of "Machine Learning" search query in Google over time

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Machine Learning'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(15, 12))
data['Machine Learning'].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Goolge Searches for Machine Learning', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```

## Conclusion

From the graph, we can see that there is a huge spike of increase in the searches about "Machine Learning" on Google in the year 2022. This is how the Google Search Analysis can be perfomed based on any keyword.
